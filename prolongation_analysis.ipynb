```python
# импортируем библиотеки
import pandas as pd
```

```python
# загружаем таблицы
financial_df = pd.read_csv('/content/financial_data.csv')
prolongations_df = pd.read_csv('/content/prolongations.csv')
```

```python
financial_df.head()
```

```python
prolongations_df.head()
```

```python
# почистим таблицы
financial_df.columns = financial_df.columns.str.lower().str.replace(' ', '_')
prolongations_df.columns = prolongations_df.columns.str.lower()

# найдем в financial_df все колонки c отгрузками
month_cols = [col for col in financial_df.columns if '202' in col]

# приведем в числовой вид
for col in month_cols:
    s = financial_df[col].astype(str).str.lower()

    # сначала убирем мусор из чисел
    s = s.str.replace(r'\s+', '', regex=True)
    s = s.str.replace(',', '.')

    # заменим слова на числа
    s = s.replace({
    'стоп': '-1',
    'end': '-1',
    'вноль': '0',
    'nan': '0'
    })

    financial_df[col] = pd.to_numeric(s, errors='coerce').fillna(0)

financial_df.head()
```

```python
# группируем по id и суммируем все денежные колонки
financial_clean_df = financial_df.groupby('id')[month_cols].sum().reset_index()
```

```python
financial_clean_df
```

```python
# соединяем таблицы
df_merged = pd.merge(prolongations_df, financial_clean_df, on='id', how='left')

df_merged
```

```python
# в колонке 'month' уберем пробелы
df_merged['month'] = df_merged['month'].str.replace(' ', '_')

# найдем id проектов, которые надо исключить
ids_to_drop = set()
for index, row in df_merged.iterrows():
    last_month = row['month']
    last_month_index = month_cols.index(last_month)
    # проверяем все месяцы до "последнего месяца реализации" включительно
    for i in range(last_month_index + 1):
        month_to_check = month_cols[i]
        if row[month_to_check] < 0:
            ids_to_drop.add(row['id'])
            break # достаточно одного "стопа", чтобы исключить проект

# фильтруем основной датафрейм, оставляя только "успешные" проекты
df_final = df_merged[~df_merged['id'].isin(ids_to_drop)].copy()

df_final
```

```python
# получаем список уникальных менеджеров
managers = df_final['am'].unique()
# добавляем "весь отдел", чтобы посчитать и общие цифры
managers_and_total = list(managers) + ['Весь отдел']

# список для хранения наших результатов
results = []

# отбираем только месяцы 2023 года
months_2023 = [col for col in month_cols if '2023' in col]

# проходим по каждому менеджеру
for manager in managers_and_total:

    # если это "весь отдел", берем всю таблицу. иначе - фильтруем по имени.
    if manager == 'Весь отдел':
        manager_df = df_final
    else:
        manager_df = df_final[df_final['am'] == manager]

    # теперь проходим по месяцам 2023 года, начиная с марта
    for i in range(2, len(months_2023)):
        # определяем наши месяцы для расчета
        current_month = months_2023[i]
        prev_month = months_2023[i-1]
        two_months_ago = months_2023[i-2]

        # --- считаем к1 (продление сразу) ---
        # 1. находим проекты, которые закончились в прошлом месяце
        ended_prev = manager_df[manager_df['month'] == prev_month]
        # 2. считаем, сколько они заплатили в свой последний месяц (знаменатель)
        base_k1 = ended_prev[prev_month].sum()
        # 3. считаем, сколько они же заплатили в текущем месяце (числитель)
        prolonged_k1 = ended_prev[current_month].sum()
        # 4. считаем коэффициент
        k1 = prolonged_k1 / base_k1 if base_k1 > 0 else 0

        # --- считаем к2 (через месяц продлили) ---
        # 1. находим проекты, которые закончились 2 месяца назад
        ended_2_ago = manager_df[manager_df['month'] == two_months_ago]
        # 2. из них выбираем тех, кто не платил в прошлом месяце
        not_prolonged = ended_2_ago[ended_2_ago[prev_month] <= 0]
        # 3. считаем, сколько они заплатили в свой последний месяц (знаменатель)
        base_k2 = not_prolonged[two_months_ago].sum()
        # 4. считаем, сколько они же заплатили в текущем месяце (числитель)
        prolonged_k2 = not_prolonged[current_month].sum()
        # 5. считаем коэффициент
        k2 = prolonged_k2 / base_k2 if base_k2 > 0 else 0

        # добавляем результат в наш список
        results.append({
            'Менеджер': manager,
            'Месяц': current_month.replace('_', ' ').capitalize(),
            'K1': k1,
            'K2': k2
        })

# превращаем список с результатами в красивую таблицу
results_df = pd.DataFrame(results)

results_df
```

```python
# создаем сводную таблицу
report_pivot = results_df.pivot_table(index='Менеджер', columns='Месяц', values=['K1', 'K2'])

# считаем средние годовые коэффициенты
yearly_report = results_df.groupby('Менеджер')[['K1', 'K2']].mean()
yearly_report.rename(columns={'K1': 'Годовой K1 (средний)', 'K2': 'Годовой K2 (средний)'}, inplace=True)

# соединяем все в один отчет
final_report = pd.concat([yearly_report, report_pivot], axis=1)

# сохраняем отчет в excel
final_report.to_excel("prolongation_report_2023.xlsx", sheet_name="Отчет по пролонгациям")

# выводим финальный отчет на экран
final_report
```
